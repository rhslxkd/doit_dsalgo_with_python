## 📖 오늘의 핵심 개념

### 검색과 키
검색 조건이 주목하는 항목을 **키(key)**라고 해. 오늘 다루는 건 **배열 검색** 중 가장 기본인 **선형 검색(순차 검색)**이야.

### 선형 검색의 종료 조건 (2가지 — 이게 핵심)
1. **검색할 값을 찾지 못하고 배열의 맨 끝을 지나간 경우** → 검색 실패
2. **검색할 값과 같은 원소를 찾는 경우** → 검색 성공

이 중 **어느 하나라도 성립하면** 스캔을 종료해. 매 반복마다 이 2개를 다 체크해야 하니까, 원소 개수가 n개면 평균 n/2번 판단하게 돼.

### 보초법(sentinel method)
매번 "배열 끝까지 갔나?"를 체크하는 비용을 줄이는 트릭. 배열 맨 끝에 **찾는 값 자체(보초)**를 미리 심어두면, "값을 못 찾고 끝까지 감"이라는 상황 자체가 사라져 — 왜냐하면 보초 덕분에 **반드시 어딘가에서 값을 찾게** 되거든. 그래서 종료 조건 1번(경계 검사)이 필요 없어지고, 판단 횟수가 절반으로 줄어.

`Muse`가 짠 `ssearch_sentinel.py`에서 `copy.deepcopy(seq)`를 쓴 이유가 바로 여기 있어 — 원본 배열을 훼손하지 않고 보초를 추가하기 위해서. (5일차 얕은/깊은 복사 개념이 여기서 실전으로 쓰이는 거야!)

### R-1. 약수의 합

```
import math

def divisor_sum(n: int) -> int:
    total = 0
    sq = int(math.isqrt(n))
    for i in range(1, sq + 1):
        if n % i == 0:
            total += i
            if i != n // i:  # 완전제곱수 짝 중복 방지
                total += n // i
    return total

print(divisor_sum(12))  # 28
print(divisor_sum(16))  # 31
```

**왜**: 약수는 √n을 기준으로 `(i, n//i)` 쌍을 이룬다는 5일차 원리 그대로야. 다른 점은 소수 판정 때는 "쌍이 하나라도 있으면 실패"만 보면 됐지만, 여기선 **쌍의 값 자체를 더해야** 하니까 `i`와 `n//i`를 각각 더해줘야 해. `i == n // i`인 경우(완전제곱수, 예: 16의 4)는 같은 수를 두 번 더하면 안 되니까 한 번만 더해.

### 3. 종료 조건 설명

- 종료 조건 1: `if i == len(a)` → 배열 끝을 지나감 (검색 실패)

**왜 하나만 성립해도 되는가**: "찾거나, 다 뒤졌는데 없거나" 둘 중 하나는 반드시 일어나야 스캔이 끝날 수 있어. 두 조건은 동시에 참일 수 없는 배타적 상황이니까, `or` 관계로 하나라도 성립하면 반복을 멈추면 돼.

### 4. 디버깅 정답 🔥

**버그**: `if idx == 1:` → **`if idx == -1:`** 이어야 해.

- **원인**: `seq_search`(선형 검색 함수)는 검색 실패 시 **`-1`**을 반환한다고 정의했는데, 테스트 코드에서는 `idx == 1`(그냥 정수 1)과 비교하고 있어.
- **영향**: `idx`가 실제로 `-1`이어도 `idx == 1`은 항상 `False`니까 `else` 분기로 빠져서 "검색값은 x[-1]에 있습니다" 같은 엉뚱한 결과가 나와. (파이썬은 `x[-1]`을 마지막 원소로 해석하니 에러도 안 나고 조용히 틀린 값을 보여줘서 더 위험해.)
- **해결**: `if idx == -1:`로 수정.
- **대안**: 매직 넘버(-1) 자체를 조건마다 손으로 안 쓰고 싶으면, `NOT_FOUND = -1` 같은 이름 붙은 상수로 빼두면 이런 오타를 원천 차단할 수 있어.

### 6. 판단 횟수 설명

기존: 매 반복마다 `if i == len(a)` + `if a[i] == key` → **2개** 체크.
보초법: `if a[i] == key` **1개**만 체크 (경계 검사 생략).

**왜 가능한가**: 보초 덕분에 "값을 못 찾고 끝까지 감"이라는 경우 자체가 사라져 — 보초 자리에 반드시 `key`와 같은 값이 있으니, `a[i] == key`는 늦어도 보초 위치에서 **반드시 참**이 돼. 그래서 경계를 넘어가는지 따로 검사할 필요가 없어져.

### 7. 보초법 구현 정답

```
from typing import Any, Sequence
import copy

def seq_search_sentinel(seq: Sequence, key: Any) -> int:
    """시퀀스 seq에서 key와 일치하는 원소를 선형 검색(보초법)"""
    a = copy.deepcopy(seq)
    a.append(key)

    i = 0
    while True:
        if a[i] == key:
            break
        i += 1
    return -1 if i == len(seq) else i

print(seq_search_sentinel([6, 4, 3, 2, 1, 2, 8], 2))    # 3
print(seq_search_sentinel([6, 4, 3, 2, 1, 2, 8], 5))    # -1
```

### 8. deepcopy 이유 설명

- **`seq` 자체에 바로 `.append(key)`했다면**: 원본 배열이 오염돼. 검색 함수를 호출했을 뿐인데 원래 배열 맨 끝에 보초 값이 남아버리는 부작용이 생겨 — 함수는 "검색만" 해야지 원본을 바꾸면 안 되니까 이건 명백한 버그야.
- **얕은 복사(`copy.copy`/`seq[:]`)를 썼다면**: `seq`의 원소가 정수·문자열처럼 **불변(immutable) 단순값**이면 사실 문제없어. 얕은 복사든 깊은 복사든 바깥 리스트가 새로 만들어지고, `.append(key)`는 그 새 바깥 리스트에만 영향을 주니까. 하지만 만약 `seq`의 원소 자체가 **리스트 같은 가변 객체**라면, 얕은 복사는 안쪽 원소를 원본과 공유하기 때문에 (R-2에서 확인한 것처럼) 나중에 안쪽 원소를 건드리는 로직이 있으면 원본이 오염될 위험이 있어. `deepcopy`를 쓰면 이런 위험을 원천 차단하니까 "안전빵"으로 깊은 복사를 선택한 거야.

---
### 9. Sequence 실험 해설

```
from typing import Any, Sequence

def seq_search(a: Sequence, key: Any) -> int:
    for i in range(len(a)):
        if a[i] == key:
            return i
    return -1

print(seq_search((4, 7, 5.6, 2, 3.14, 1), 5.6))   # 2
print(seq_search('string', 'n'))                   # 4
print(seq_search(['DTS', 'AAC', 'FLAC'], 'AAC'))   # 1
```

**왜**: 튜플, 문자열, 리스트는 전부 **시퀀스(Sequence)** — `len()`으로 길이를 잴 수 있고 `a[i]`로 인덱싱할 수 있다는 공통 인터페이스를 갖고 있어. `seq_search`는 이 두 가지 연산(`len`, `[i]`)만 사용하니까, 그 조건을 만족하는 어떤 자료형이든 그대로 동작해. 이게 타입 힌트를 구체적인 `list`가 아니라 `Sequence`로 잡은 이유야.
