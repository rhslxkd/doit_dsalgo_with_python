### 개념 6 🟡 [예측] — `break` vs `continue`

`break`는 **루프 전체 탈출**, `continue`는 **이번 회차만 건너뛰고 다음으로**. 출력을 예측해봐.

````python
# break: 5를 만나면 멈춤
for i in range(1, 10):
    if i == 5:
        break
    print(i, end=' ')
print("  <- break")

# continue: 짝수는 건너뜀
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i, end=' ')
print("  <- continue")
````

```python
def triangle_left(n):
    for i in range(n):
        for _ in range(i+1): print('*', end='')
        print()
def triangle_right(n):
    for i in range(n):
        for _ in range(n-i-1): print(' ', end='')   # 공백 ' ', end=''
        for _ in range(i+1):   print('*', end='')
        print()
```

**구현 10** 🔴
```python
def factor_pairs(area):
    for i in range(1, area+1):
        if i*i > area: break
        if area % i:   continue
        print(f"{i} * {area//i}")
```
> `i*i > area`에서 멈추는 이유: 약수는 √area 기준으로 대칭이라 그 너머는 이미 나온 쌍의 반복.
