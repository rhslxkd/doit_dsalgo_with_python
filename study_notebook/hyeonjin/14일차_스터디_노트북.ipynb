## 📦 부록 — 오늘의 코드 모음

이 셀을 **먼저 실행**해. (교재 실습 5-3, 5-4, 5-5 + Stack 클래스)

````
from typing import Any
from collections import deque
import io, contextlib

# --- 11일차 stack.py (실습 4C-1) ---
class Stack:
    '''고정 길이 스택 클래스(collections.deque를 사용)'''
    def __init__(self, maxlen: int = 256) -> None:
        self.capacity = maxlen
        self.__stk = deque([], maxlen)
    def __len__(self) -> int:
        return len(self.__stk)
    def is_empty(self) -> bool:
        return not self.__stk
    def is_full(self) -> bool:
        return len(self.__stk) == self.__stk.maxlen
    def push(self, value: Any) -> None:
        self.__stk.append(value)
    def pop(self) -> Any:
        return self.__stk.pop()
    def peek(self) -> Any:
        return self.__stk[-1]
    def dump(self):
        print(list(self.__stk))
    def to_list(self):
        return list(self.__stk)

# --- 실습 5-3: 순수한 재귀 (recur1.py) ---
def recur1(n: int) -> None:
    '''순수한 재귀 함수 recur의 구현'''
    if n > 0:
        recur1(n - 1)
        print(n, end=' ')
        recur1(n - 2)

# --- 실습 5-4: 꼬리 재귀 제거 (recur1a.py) ---
def recur1a(n: int) -> None:
    '''꼬리 재귀를 제거한 recur() 함수'''
    while n > 0:
        recur1a(n - 1)
        print(n, end=' ')
        n = n - 2

# --- 실습 5-5: 재귀 완전 제거 (recur1b.py) ---
def recur1b(n: int) -> None:
    '''재귀를 제거한 recur() 함수'''
    s = Stack(n if n > 0 else 1)
    while True:
        if n > 0:
            s.push(n)
            n = n - 1
            continue
        if not s.is_empty():
            n = s.pop()
            print(n, end=' ')
            n = n - 2
            continue
        break

# --- recur2.py: 거꾸로 출력 ---
def recur2(n: int) -> None:
    '''순수한 재귀 함수 recur의 구현(거꾸로 출력)'''
    if n > 0:
        recur2(n - 2)
        print(n, end=' ')
        recur2(n - 1)

def capture(f, n):
    '''함수 출력을 리스트로 캡처하는 헬퍼'''
    buf = io.StringIO()
    with contextlib.redirect_stdout(buf):
        f(n)
    return buf.getvalue().split()

print('준비 완료')
````

## 📖 오늘의 핵심 개념 (구조 중심)

### 순수한 재귀(genuinely recursive)
함수 안에서 **재귀 호출을 2번 이상** 하는 것.

| | `factorial`, `gcd` (13일차) | `recur` (오늘) |
|---|---|---|
| 재귀 호출 횟수 | 1번 | **2번** |
| 호출 구조 | **일직선** (사슬) | **나무(트리)** |
| 동작 예측 | 쉬움 | 어려움 |

이 "1번 vs 2번"이 오늘 모든 것의 출발점이야. 호출이 2개로 갈라지니까 구조가 **트리**가 되고, 그래서 분석 방법이 따로 필요해진 거야.

### 분석 방법 1 — 하향식 분석(top-down analysis)
**가장 위의 호출부터 시작해서 계단식으로 파고 들어가는** 방법. 교재 193p 그림 5-5.

```
recur(4)
├─ recur(3)  ← 먼저 이걸 알아야
│   ├─ recur(2)
│   │   ├─ recur(1) ...
```

한 단계를 **완전히 끝내야** 위로 올라갈 수 있어.

> ⚠️ **하향식의 함정**: 트리를 내려가다 보면 `recur(1)`, `recur(2)` 같은 **같은 함수를 여러 번** 호출하게 돼. 그래서 교재 194p는 "하향식 방식이 반드시 효율적이라고 말할 수는 없습니다"라고 해. (4번 문제에서 직접 세어볼 거야)

### 분석 방법 2 — 상향식 분석(bottom-up analysis)
**아래(작은 값)부터 쌓아 올리는** 방법. 교재 195p 그림 5-6.

```
recur(-1) : 아무것도 안 함
recur(0)  : 아무것도 안 함
recur(1)  : recur(0) → 1 출력 → recur(-1)  =  "1"
recur(2)  : recur(1) → 2 출력 → recur(0)   =  "1 2"
recur(3)  : recur(2) → 3 출력 → recur(1)   =  "1 2 3 1"
recur(4)  : recur(3) → 4 출력 → recur(2)   =  "1 2 3 1 4 1 2"
```

작은 것부터 답을 구해서 **재사용**하니까 중복이 없어. (이게 나중에 배울 "동적 계획법(DP)"의 씨앗이야)

**핵심**: 하향식과 상향식은 **결과가 같고 사고 방향만 반대**야. 위에서 파고들 것인가, 아래서 쌓아 올릴 것인가.

### 재귀 제거 1단계 — 꼬리 재귀(tail recursion) 제거
**꼬리 재귀** = 함수의 **맨 끝**에서 하는 재귀 호출. `recur`에선 `recur(n - 2)`가 여기 해당.

맨 끝 호출은 "n을 n-2로 바꾸고 함수 처음으로 돌아가라"와 **완전히 같은 의미**야. 뒤에 할 일이 아무것도 없으니까!

```python
# 재귀                        # 꼬리 재귀 제거
if n > 0:                     while n > 0:
    recur(n - 1)                  recur(n - 1)
    print(n)                      print(n)
    recur(n - 2)   ← 꼬리          n = n - 2      ← 그냥 값 바꾸고 루프
```
`if` → `while`, 마지막 재귀 호출 → `n = n - 2`. 이게 전부야.

### 재귀 제거 2단계 — 맨 앞 재귀는 스택이 필요
그런데 **맨 앞의 `recur(n - 1)`은 같은 방법으로 못 없애.**

**왜?** `n`을 출력하려면 `recur(n-1)`이 **완전히 끝난 뒤**여야 하는데, `n`을 `n-1`로 바꿔버리면 **원래 n값을 잃어버려**. 그래서 원래 n을 **어딘가에 저장**해뒀다가 나중에 꺼내 써야 해.

→ 저장했다가 **역순으로** 꺼내야 하니까 → **스택!** (11일차 회수)

```python
def recur1b(n):
    s = Stack(n)
    while True:
        if n > 0:
            s.push(n)      # ① 원래 n을 저장
            n = n - 1      # ② 더 깊이 내려감
            continue
        if not s.is_empty():
            n = s.pop()    # ④ 저장했던 값 꺼내고
            print(n)       # ⑤ 출력
            n = n - 2      # ⑥ 꼬리 재귀 부분
            continue
        break              # 스택도 비고 n도 0 이하면 끝
```

**구조 요약**: `push` = 재귀의 "내려가기", `pop` = "올라오기". 13일차 9번에서 factorial로 해본 것과 같은 원리인데, 여기선 갈래가 2개라 훨씬 흥미로워.

### 4. [구조] 하향식 vs 상향식 — 왜 두 가지 분석법이 있나 🔥🔥

교재 194p: "꼭대기부터 분석하면 이렇게 같은 함수를 여러 번 호출할 수 있으므로 **하향식 방식이 반드시 효율적이라고 말할 수는 없습니다.**"

이 문장이 무슨 뜻인지 **숫자로** 확인해보자.

````
from collections import Counter

calls = Counter()
def recur_counted(n):
    calls[n] += 1          # 이 값으로 호출된 횟수 기록
    if n > 0:
        recur_counted(n - 1)
        recur_counted(n - 2)

calls.clear()
recur_counted(4)
print('=== recur(4)를 하향식으로 실행할 때 호출 횟수 ===')
for k in sorted(calls, reverse=True):
    bar = '█' * calls[k]
    print(f'  recur({k:2}) : {calls[k]}번  {bar}')
print(f'  ------------------------')
print(f'  총 호출 횟수: {sum(calls.values())}번')
print()
print('상향식이라면? recur(-1)~recur(4) 각각 한 번씩만 계산 = 6번이면 충분!')
````

**(a)** `recur(4)` 하나를 실행하는데 `recur(0)`이 몇 번이나 중복 호출됐어? 왜 이렇게 중복될까? (힌트: 트리에서 `recur(0)`에 도달하는 경로가 여러 개)

**(b)** 상향식으로 하면 왜 중복이 없을까? 무엇을 "재사용"하는 거지?

**(c)** 그럼 하향식 분석은 쓸모없는 걸까? 하향식이 더 자연스러운 상황은 언제일까?

### 5. [구조] 왜 꼬리 재귀만 쉽게 제거될까 🔥🔥

교재 196~197p의 핵심 질문이야. 두 재귀 호출을 비교해봐:

```python
if n > 0:
    recur(n - 1)    # ← 맨 앞: 제거 어려움
    print(n)
    recur(n - 2)    # ← 맨 끝(꼬리): 제거 쉬움
```

교재 197p는 맨 앞 재귀에 대해 이렇게 말해:
> "n값을 출력하기 전에 recur(n − 1)을 실행해야 하기 때문입니다. (...) 현재의 n값을 임시로 저장할 필요가 있기 때문입니다."

아래 실험으로 **"n을 잃어버린다"**가 무슨 뜻인지 직접 봐.

````
print('=== 맨 앞 재귀를 꼬리 재귀처럼 없애보려는 (실패하는) 시도 ===')
print()

def broken_recur(n):
    '''맨 앞 재귀를 n = n-1 로 그냥 바꿔버린 잘못된 버전'''
    while n > 0:
        n = n - 1        # recur(n-1)을 이렇게 대체하면?
        print(n, end=' ')  # 근데 여기서 출력할 n은... 원래 n이어야 하는데!
        n = n - 2

print('원본 recur1(4)   :', ' '.join(capture(recur1, 4)))
print('망가진 버전(4)   :', ' '.join(capture(broken_recur, 4)))
print()
print('-> n을 먼저 바꿔버리니 "출력해야 할 원래 n"이 사라졌다!')
````

### 9. [구조] `Stack(n)` 크기는 왜 딱 `n`일까

`recur1b`는 스택을 `Stack(n)`으로 만들어. 왜 하필 `n`일까? 더 크거나 작으면 안 될까?

**주의**: 11일차에서 봤듯이 이 `Stack`은 `collections.deque(maxlen=...)` 기반이라, **가득 찬 상태에서 push하면 예외 없이 반대쪽을 조용히 밀어내.**

````
def max_stack_depth(n):
    '''recur1b 실행 중 스택이 최대 몇 개까지 쌓이는지'''
    s = Stack(10000)   # 넉넉하게
    maxd = 0
    while True:
        if n > 0:
            s.push(n); maxd = max(maxd, len(s)); n -= 1; continue
        if not s.is_empty():
            n = s.pop(); n -= 2; continue
        break
    return maxd

print('=== 최대 스택 깊이 vs Stack(n) 용량 ===')
for n in [1, 2, 3, 4, 5, 10, 20]:
    # 주의: n을 더 키우면 7번에서 본 지수 시간 때문에 셀이 안 끝나!
    d = max_stack_depth(n)
    print(f'  n={n:3}: 최대 깊이={d:3}, 용량={n:3}  →  {"딱 맞음" if d == n else "다름!"}')

print()
print('=== 만약 용량이 부족하면? (deque는 조용히 밀어냄) ===')
def recur1b_small_stack(n, cap):
    s = Stack(cap)
    orig_n = n
    while True:
        if n > 0:
            s.push(n); n -= 1; continue
        if not s.is_empty():
            n = s.pop(); print(n, end=' '); n -= 2; continue
        break

print('정상 (용량 4):', ' '.join(capture(lambda n: recur1b_small_stack(n, 4), 4)))
print('부족 (용량 2):', ' '.join(capture(lambda n: recur1b_small_stack(n, 2), 4)))
print()
print('-> 용량이 부족해도 에러가 안 나고 조용히 틀린 결과가 나온다!')
````

**(b)** 용량이 부족할 때 `RecursionError` 같은 명확한 에러가 아니라 **조용히 틀린 답**이 나오는 게 왜 위험할까? (13일차 4번, 9일차 오픈해시 버그와 같은 결)

## ✅ 정답 & 해설

### R-1. `factorial` vs `recur` 구조

- `factorial`은 재귀 호출 **1번**, `recur`은 **2번**(`recur(n-1)`, `recur(n-2)`).
- **왜 구분이 필요한가**: 호출이 1번이면 실행 흐름이 **일직선**이라 "n → n-1 → n-2 → ... → 0" 하나의 사슬로 예측 가능해. 2번이면 매 단계마다 **두 갈래로 갈라져서** 전체 구조가 **트리(나무)**가 돼. 그래서 교재 193p처럼 "n이 3이나 5라면 어떤 결과를 출력할지는 간단히 알 수 없습니다"가 되는 거야. 분석 도구(하향식/상향식)가 따로 필요해진 이유.
- **모양**: `factorial(4)`는 **직선**(4→3→2→1→0), `recur(4)`는 **트리**(교재 그림 5-5처럼 아래로 갈수록 넓게 퍼짐).

---
### R-2. `recur`의 기저 조건

- 기저 조건은 **`if n > 0`의 "거짓" 쪽**이야. `else`가 없으니 `n <= 0`이면 **함수가 아무것도 안 하고 그냥 끝남**.
- `recur(0)`, `recur(-1)`은 호출은 되지만 `if` 조건이 거짓이라 즉시 반환. 이게 교재 그림 5-5의 **`-`로 표시된 빈 상자**야.
- 13일차 `factorial`은 `return 1`이라는 **값을 돌려주는** 기저 조건이었는데, `recur`은 **아무 일도 안 하는 것**이 기저 조건. 반환값이 없는(`print`만 하는) 함수라서 이런 형태가 가능해.

---
### 1. `recur(3)` 추적 정답

| 호출 | 출력 |
|---|---|
| `recur(1)` | `1` |
| `recur(2)` | `1 2` |
| `recur(3)` | **`1 2 3 1`** |

`recur(3)` = `recur(2)`의 출력(`1 2`) + `3` + `recur(1)`의 출력(`1`) = `1 2 3 1`

### 2. `recur2`는 뭐가 바뀐 걸까

**결과**: `recur1(4)` → `1 2 3 1 4 1 2` / `recur2(4)` → `2 1 4 1 3 2 1` (교재 195p와 일치)

**구조적 이유**: `print(n)`은 그대로인데, **그 앞뒤에 오는 재귀 호출이 서로 바뀌었어.**
- `recur1`: "먼저 `n-1`쪽을 다 처리 → n 출력 → `n-2`쪽 처리"
- `recur2`: "먼저 `n-2`쪽을 다 처리 → n 출력 → `n-1`쪽 처리"

트리에서 **어느 가지를 먼저 타고 내려가느냐**가 바뀐 거야. 출력은 "왼쪽 가지를 다 끝낸 뒤"에 일어나니까, 왼쪽에 뭐가 오느냐가 전체 출력 순서를 완전히 뒤바꿔.

**🎯 그런데 놀랍게도, `recur2(n)`은 정확히 `recur1(n)`의 출력을 뒤집은 것과 같아** — 우연이 아니라 **모든 n에서 항상** 성립해. 왜 그런지 구조로 증명할 수 있어:

`recur1(n)`의 출력 = `recur1(n-1)` + `[n]` + `recur1(n-2)`
`recur2(n)`의 출력 = `recur2(n-2)` + `[n]` + `recur2(n-1)`

이제 `recur1(n)`을 통째로 뒤집어보면 — 리스트를 뒤집으면 **순서가 반대가 되고 각 조각도 뒤집혀**:

reverse(`recur1(n)`) = reverse(`recur1(n-2)`) + `[n]` + reverse(`recur1(n-1)`)

여기서 더 작은 값들에 대해 "뒤집으면 recur2가 된다"가 이미 참이라고 하면(귀납 가정):

= `recur2(n-2)` + `[n]` + `recur2(n-1)` = **`recur2(n)`** ✓

즉 `recur2`는 `recur1`의 **거울상(mirror)** 구조야. 두 재귀 호출의 순서를 바꾸는 것 = 트리의 좌우를 뒤집는 것 = 출력 순서가 완전히 역전되는 것. 교재가 이 버전을 "**거꾸로 출력**"이라고 부른 이유가 바로 이거야.

### 4. 하향식 vs 상향식 🔥🔥

**실행 결과**: `recur(4)` 총 **15번** 호출. 그중 `recur(0)`이 **5번**, `recur(1)`이 **3번**, `recur(-1)`이 **3번** 중복!

**(a)** 트리를 그려보면 `recur(0)`에 도달하는 **경로가 여러 개**야. `recur(4)→recur(3)→recur(2)→recur(1)→recur(0)`으로도 가고, `recur(4)→recur(2)→recur(0)`으로도 가고... 각 경로마다 **똑같은 계산을 처음부터 다시** 해. 이미 답을 아는데도 저장을 안 해두니까.

**(b)** 상향식은 `recur(-1)`, `recur(0)`, `recur(1)`, ... 순서로 **작은 것부터 한 번씩만** 구하고, 그 결과를 위 단계에서 **재사용**해. `recur(4)`를 구할 땐 이미 알고 있는 `recur(3)`과 `recur(2)`의 답을 **가져다 쓰기만** 하면 되니까 중복이 0. → 6번의 계산으로 끝. (15번 vs 6번!)

**(c)** 하향식도 충분히 쓸모 있어. **목적이 달라**:
- **하향식**: "이 함수가 **어떤 순서로 동작**하는가"를 이해할 때. 실행 흐름을 그대로 따라가니까 디버깅·추적에 자연스러워. (교재 그림 5-5)
- **상향식**: "**최종 결과가 뭔가**"를 효율적으로 구할 때. 중복 없이 답을 쌓아 올려. (교재 그림 5-6)

즉 하향식은 **이해·추적의 도구**, 상향식은 **계산·최적화의 도구**야. 참고로 상향식으로 결과를 저장해두고 재사용하는 아이디어가 나중에 배울 **동적 계획법(DP)** 이고, 하향식에 저장을 붙이면 **메모이제이션**이 돼.

### 5. 왜 꼬리 재귀만 쉽게 제거될까 🔥🔥

**(a) 꼬리 재귀가 쉬운 이유**: `recur(n-2)`는 함수의 **맨 마지막 줄**이라, 그 호출이 끝나면 **함수도 그냥 끝나.** 즉 "돌아와서 마저 할 일"이 **없어**. 그러니 "새 n값으로 함수를 처음부터 다시 실행"하는 것과 완전히 같고, 그건 `n = n - 2` + `while` 루프로 표현할 수 있어.

**(b) 맨 앞 재귀가 어려운 이유**: `recur(n-1)` 뒤에는 **`print(n)`과 `recur(n-2)`가 남아있어.** 그런데 `n = n - 1`로 값을 덮어써 버리면, 나중에 출력해야 할 **원래 n을 영영 잃어버려.** (실험에서 `broken_recur(4)`가 `3 0` 같은 엉뚱한 값을 낸 이유가 정확히 이거야)

**(c)** 저장한 값은 **가장 최근에 저장한 것부터** 꺼내 써야 해. `recur(4)`는 `recur(3)`을, `recur(3)`은 `recur(2)`를... 안쪽으로 파고들었다가 **가장 안쪽부터 되짚어 나오니까**. 이건 **LIFO** → **스택**이 정답. (11일차 회수)

### 6. 스택 변화 추적 🔥🔥

**(a) 순서**: 푸시는 `4, 3, 2, 1` 순서로 쌓이고, 팝은 **꼭대기부터** `1` → (n=1-2=-1이라 다시 팝) `2` → ... 이렇게 **역순**으로 나와. LIFO니까. 다만 팝한 뒤 `n = n - 2`를 하고 그 값이 0보다 크면 **다시 푸시**가 일어나서, 단순 역순이 아니라 푸시/팝이 섞여.

**(b)** `push` = 재귀의 **"내려가기"**(더 작은 문제로 파고듦, `recur(n-1)` 호출에 해당). `pop` = 재귀의 **"올라오기"**(안쪽이 끝나서 되돌아옴 → 미뤄뒀던 `print(n)` 실행).

**(c)** 팝한 뒤의 `n = n - 2`는 **꼬리 재귀 `recur(n-2)`**에 해당해. 이 값이 0보다 크면 그 서브트리를 또 탐색해야 하므로 다시 푸시 루프로 들어가. 즉 **하나의 while 루프가 트리 전체를 순회**하는 구조야.

---
### 7. 등가성 & 출력 개수 패턴

**등가성**: n=0~8 전부 세 버전 일치 ✓ (재귀 제거가 의미를 바꾸지 않았다는 증명)

**(a) 점화식**: `f(n) = f(n-1) + 1 + f(n-2)` (단, `f(n) = 0` for n ≤ 0)
- `recur(n-1)`이 출력하는 개수 + `n` 자기 자신 1개 + `recur(n-2)`가 출력하는 개수

**(b)** **피보나치**와 거의 같아! 실제로 `f(n) = Fib(n+2) - 1`이 성립해.
- f(1)=1=Fib(3)-1, f(2)=2=Fib(4)-1, f(3)=4=Fib(5)-1, f(4)=7=Fib(6)-1 ...

**(c)** 피보나치 수열은 **지수적으로** 증가해 (대략 황금비 φ≈1.618의 거듭제곱). 즉 `recur(n)`의 시간 복잡도는 **O(φⁿ) ≈ O(1.618ⁿ)** — 지수 시간! n을 1 늘리면 대략 **1.6배**씩 늘어나. n=40쯤 되면 출력이 수억 개가 돼서 사실상 못 돌려. 이게 "순수한 재귀"가 위험할 수 있는 이유이자, 4번에서 본 **중복 호출**의 대가야.

---
### 8. `recur2` 재귀 제거 정답
```python
def recur2_no_recursion(n):
    s = Stack(n if n > 0 else 1)
    while True:
        if n > 0:
            s.push(n)
            n = n - 2      # recur2의 맨 앞 재귀가 recur(n-2)니까
            continue
        if not s.is_empty():
            n = s.pop()
            print(n, end=' ')
            n = n - 1      # recur2의 꼬리 재귀가 recur(n-1)니까
            continue
        break
```
**핵심**: `recur1b`와 **숫자 2와 1의 위치만 서로 바뀜.** 구조(푸시→내려가기, 팝→출력→꼬리처리)는 완전히 똑같아. 이게 "재귀 제거"라는 기법이 **특정 함수에만 통하는 요령이 아니라 일반적인 패턴**이라는 증거야.

---
### 9. `Stack(n)` 크기

**(a)** `recur1b`는 처음에 `n > 0`인 동안 계속 푸시만 해: `n, n-1, n-2, ..., 2, 1` → 정확히 **n개**. 그 뒤론 팝이 섞이면서 이보다 더 높이 쌓이는 일이 없어(팝 후 `n-2`는 항상 방금 팝한 값보다 작으니까). 그래서 `Stack(n)`이 **딱 필요한 만큼**이야. 낭비도 부족함도 없는 정확한 설계.

**(b)** 실험에서 봤듯이 용량이 부족하면 `deque(maxlen)`이 **바닥 쪽 원소를 조용히 버려서**, 에러 없이 **틀린 결과**가 나와. 크래시는 즉시 발견되지만 조용한 실패는 "결과가 좀 이상한데?" 하고 넘어가기 쉬워서 훨씬 위험해. (13일차 `factorial(-1)`이 조용히 1을 반환한 것, 9일차 오픈해시가 `None` 값을 "키 없음"으로 착각한 것과 **완전히 같은 종류의 위험**)

**(c)** 11일차 `FixedStack`은 가득 찬 상태에서 push하면 **`Full` 예외**를 던져. 그러면 "용량이 부족하다"는 걸 **즉시, 명확히** 알 수 있어서 조용히 틀린 답이 나오는 상황을 막을 수 있어. 11일차 7번에서 다룬 "가득 참을 오류로 볼지 정상으로 볼지"의 선택이 여기서 실제 결과를 가르는 셈이야.

## 📌 핵심 3줄 요약

1. **순수한 재귀**(재귀 호출 2번 이상)는 구조가 **트리**라 예측이 어렵고, 그래서 **하향식**(동작 순서 추적용)과 **상향식**(결과 계산용) 두 가지 분석법이 필요하다.
2. **꼬리 재귀는 쉽게 제거된다** — 뒤에 할 일이 없으니 `n = n - 2` + `while`로 충분. **맨 앞 재귀는 원래 n을 잃어버려서** 제거가 어렵고, 저장했다 역순으로 꺼내야 하니 **스택**이 필요하다.
3. 재귀 제거의 구조는 항상 같다: **`push` = 내려가기, `pop` = 올라오기**. `recur1`이든 `recur2`든 숫자만 바뀔 뿐 패턴은 동일하다.
