## 📖 오늘의 핵심 개념

### 재귀(recursion)
어떤 대상을 정의할 때 **자기 자신을 포함해서** 정의하는 것. 예: 자연수의 정의
- 1은 자연수다.
- 어떤 자연수의 바로 다음 수도 자연수다.

이렇게 무한히 존재하는 대상을 **재귀적 정의(recursive definition)**로 유한한 규칙 몇 개로 표현할 수 있어.

### 팩토리얼(factorial)의 재귀적 정의
```
0! = 1
n > 0이면  n! = n × (n-1)!
```
- **기저 조건(base case)**: `0! = 1` — 더 이상 재귀하지 않고 즉시 답이 나오는 지점
- **재귀 단계(recursive step)**: `n! = n × (n-1)!` — 더 작은 문제로 자신을 다시 호출

기저 조건이 없으면? → **무한 재귀 → RecursionError** (R-2에서 확인한 것과 같은 현상)

### 재귀 호출(recursive call)과 콜 스택
`factorial(3)`을 호출하면:
1. **내려가기**: `factorial(3)`은 `factorial(2)`의 결과가 필요해서 호출 → 그 안에서 또 `factorial(1)` 호출 → `factorial(0)` 호출. 각 호출은 "아직 안 끝난 채" **스택에 쌓임**.
2. **바닥 도달**: `factorial(0)`이 `n>0`이 아니므로 즉시 `1` 반환 (기저 조건)
3. **올라오기**: 쌓여있던 호출들이 **역순으로(LIFO)** 하나씩 풀리며 곱셈을 완료

이게 11일차 스택의 push(내려가기)/pop(올라오기)과 정확히 같은 구조. 파이썬 인터프리터가 자동으로 이 스택 관리를 해줘서 우리가 직접 push/pop을 안 부르는 것뿐.

### 직접 재귀 vs 간접 재귀
- **직접 재귀(direct recursion)**: 함수가 **자기 자신**을 직접 호출. (`factorial`이 `factorial`을 호출)
- **간접 재귀(indirect recursion)**: `a()`가 `b()`를 호출하고, `b()`가 다시 `a()`를 호출하는 식으로 **여러 함수를 거쳐** 돌아옴.

### 유클리드 호제법(Euclidean algorithm) — 최대공약수(GCD)
두 정수 x, y의 최대공약수를 재귀로 구하는 유명한 알고리즘:
```
y == 0 이면    gcd(x, y) = x
y != 0 이면    gcd(x, y) = gcd(y, x % y)
```
기하학적으로는: 가로세로 x, y인 직사각형을 정사각형으로 채워나가다 보면, 마지막에 남는 정사각형의 한 변 길이가 최대공약수 (교재 189p 그림 5-4).

### math 모듈의 기존 함수 활용
- `math.factorial(x)`: 팩토리얼 (x가 음수면 `ValueError`)
- `math.gcd(a, b)`: 최대공약수 (`gcd(0, 0)`은 `0` 반환)

> 교재 189p 각주: "factorial() 함수는 재귀 함수로 정의하지 않는 것이 오히려 간단하고 효율적입니다." — 재귀는 **개념 이해용 예제**로는 좋지만, 실전에선 반복문이나 내장 함수가 더 나을 때가 많아.

|---|---|---|
| 1 | factorial(4) 호출 | [× 4] |

| 8 | factorial(3) 재개 → ? 반환 | |

````
def factorial(n, depth=0):
    indent = '  ' * depth
    print(f'{indent}▶ factorial({n}) 호출')
    if n > 0:
        result = n * factorial(n - 1, depth + 1)
        print(f'{indent}◀ factorial({n}) 재개, {n} * factorial({n-1})의 결과로 {result} 반환')
        return result
    else:
        print(f'{indent}◀ factorial(0) -> 즉시 1 반환')
        return 1

print('결과:', factorial(4))
````

### 5. [예측] 재귀 vs 반복문 — 같은 결과, 다른 방식

`factorial`을 반복문(for/while)으로도 짤 수 있어. 아래 두 버전이 같은 값을 내놓는지 확인하고, **스택을 쓰는지 여부**를 설명해봐.

````
def factorial_recursive(n):
    if n > 0:
        return n * factorial_recursive(n - 1)
    else:
        return 1

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

for n in [0, 1, 5, 10]:
    r1 = factorial_recursive(n)
    r2 = factorial_iterative(n)
    print(f'n={n}: 재귀={r1}, 반복={r2}, 일치={r1==r2}')

print()
import sys
print('재귀 버전은 n=5000에서:', end=' ')
try:
    factorial_recursive(5000)
    print('성공')
except RecursionError:
    print('RecursionError!')

print('반복 버전은 n=5000에서:', end=' ')
try:
    factorial_iterative(5000)
    print('성공 (스택을 안 쓰니까 문제없음)')
except RecursionError:
    print('RecursionError!')
````

**(a)** 왜 반복문 버전은 n=5000에서도 멀쩡할까? (콜 스택 관점에서 설명)

**(b)** 그럼 재귀는 왜 쓰는 걸까? 교재 189p 각주("factorial은 재귀로 안 짜는 게 오히려 간단하고 효율적")를 참고해서, 재귀가 정말 유리한 경우는 뭐가 있을지 생각해봐.

### 9. [설명] 재귀 vs 스택 — 팩토리얼을 명시적 스택으로 짜면?

재귀는 사실 "언어가 대신 스택 관리를 해주는 것"이라고 했지. 그럼 11일차 `FixedStack`을 **직접 써서** 재귀 없이 factorial을 흉내낼 수 있을까?

아래 코드를 보고 이게 왜 "재귀를 스택으로 흉내낸 것"인지 설명해봐.

````
def factorial_with_explicit_stack(n):
    stack = []          # 우리가 직접 관리하는 스택
    
    # 내려가기: n, n-1, ..., 1을 스택에 쌓음
    current = n
    while current > 0:
        stack.append(current)
        current -= 1
    
    # 올라오기: 스택에서 하나씩 꺼내며 곱함
    result = 1
    while stack:
        result *= stack.pop()
    
    return result

for n in [0, 1, 5, 10]:
    print(f'{n}! = {factorial_with_explicit_stack(n)}')
````

## ✅ 정답 & 해설

### R-1. 재귀와 스택

- **LIFO**야. 가장 나중에 호출된 `factorial(0)`이 가장 먼저 반환되고, 가장 먼저 호출된 `factorial(3)`이 가장 마지막에 반환돼 — 11일차 스택의 push/pop 순서와 정확히 같아.
- **"자신"이 아니라 "자신과 똑같은 함수"**인 이유: 스택에 쌓인 `factorial(3)`, `factorial(2)`, `factorial(1)`, `factorial(0)`은 **코드는 똑같지만 각각 다른 `n`값을 가진 독립된 실행**이야. 만약 "진짜 자기 자신"이라고 생각하면 이 넷을 구별할 방법이 없어져 헷갈려. 스택의 각 프레임(칸)이 자기만의 `n`을 따로 갖고 있다고 이해해야 정확해.

`factorial(5000)`은 5000단계를 내려가야 하는데, 파이썬 기본 재귀 한계(1000)를 넘어서 **`RecursionError`**가 발생해. 이건 11일차 `FixedStack`이 `capacity`를 넘으면 `Full` 예외를 던진 것과 **근본적으로 같은 현상** — 콜 스택도 결국 유한한 메모리 공간을 쓰는 스택이라 무한정 쌓을 수 없어.

### 2. gcd 구현 정답
```python
def my_gcd(x, y):
    if y == 0:
        return x
    else:
        return my_gcd(y, x % y)
```

### 3. 기저 조건이 왜 필수인가

- 기저 조건이 없으면 `n`이 계속 작아지기만 하고 **절대 멈추지 않아** → 결국 콜 스택이 한계를 넘어 `RecursionError` (R-2에서 실제로 확인함). 무한 반복문에 탈출 조건이 없는 것과 같은 문제.
- `gcd`의 `y == 0`이 기저 조건인 이유: 유클리드 호제법은 "큰 직사각형을 정사각형으로 채우다 보면 결국 완전히 정사각형들로만 나뉘는 순간이 온다"는 원리(교재 189p). 나머지가 0이 된다는 건 "더 이상 나눌 필요 없이 완전히 딱 떨어졌다" = "정사각형만 남았다"는 뜻이라, 그 순간의 값(x)이 곧 최대공약수야.

### 4. factorial.py의 조용한 버그 🔥🔥

**(a)** `factorial(-1)` 추적:
- `n=-1`: `n > 0`이 거짓 → `else: return 1` **즉시 실행**
- 즉, 음수는 `n>0` 조건에 아예 안 걸려서 **기저 조건과 똑같이 취급**돼버려. 원래 기저 조건은 "n==0일 때만" 이어야 하는데, 코드가 `else`(= n이 0 이하 전부)로 뭉뚱그려놔서 음수도 전부 즉시 1을 반환.

**(b)** **일반 버그보다 위험한 이유**: 크래시가 나는 버그는 실행하자마자 눈에 띄어서 바로 고치게 돼. 근데 이건 **에러 없이 그럴듯한 숫자(1)를 뱉어서**, 팀원이 실수로 `factorial(-3)`을 호출해도 "어? 이상하다"를 못 느끼고 **`1`이라는 틀린 값을 다음 계산에 그대로 써버릴 수 있어.** 조용한 실패(silent failure)가 항상 더 위험한 이유 — 9일차 오픈해시에서 `value=None`을 "키 없음"으로 착각했던 버그랑 같은 종류.

**(c)** `factorial_ve.py`:
```python
if n > 0:
    return n * factorial(n - 1)
elif n == 0:
    return 1
else:
    raise ValueError
```
`n==0`을 **명시적으로 분리**해서, 진짜 기저 조건(0)과 잘못된 입력(음수)을 구별해. 만약 `elif n==0` 없이 `else: raise ValueError`만 있었다면 `n==0`도 `else`로 빠져서 **정상 입력인데 에러가 나는** 정반대 버그가 생겨. 즉 3방향 분기(`n>0` / `n==0` / `그 외`)가 정확히 필요한 이유가 있는 거야.

### 5. 재귀 vs 반복

**(a)** 반복문 버전은 **함수를 다시 호출하지 않고** `for` 루프 안에서 변수 `result`만 계속 갱신해. 콜 스택에 새 프레임이 쌓이는 게 아니라 **같은 프레임 안에서 반복**하니까, n이 아무리 커도 스택 깊이는 그대로(1단계)야. 그래서 RecursionError가 안 남.

**(b)** factorial처럼 "그냥 순서대로 곱하기/더하기"는 사실 반복문이 더 간단하고 안전해(교재 189p 각주와 동일한 결론). **재귀가 진짜 유리한 경우**는 문제 자체가 "더 작은 같은 모양의 문제로 쪼개지는" 구조일 때 — 예를 들어 트리 구조 순회(자식이 다시 트리), 하노이의 탑(다음 시간!), 퀵/병합 정렬(6장 예고편) 같은 건 반복문으로 짜면 오히려 훨씬 복잡해져. "선형으로 계속 작아지기만 하는" 문제는 반복문, "여러 갈래로 쪼개지는" 문제는 재귀가 자연스러워.

### 6. 직접 vs 간접 재귀

- `is_even`과 `is_odd`는 **간접 재귀**야 (`is_even → is_odd → is_even → ...`).
- `is_even(4)`의 콜 스택: `is_even(4) → is_odd(3) → is_even(2) → is_odd(1) → is_even(0)` → `is_even(0)`이 `True` 반환 → 역순으로 풀림.
- **디버깅이 더 어려운 이유**: 직접 재귀는 코드 한 곳(`factorial` 정의 하나)만 보면 전체 흐름을 그릴 수 있어. 간접 재귀는 `is_even`을 보다가 `is_odd`로, 다시 `is_even`으로 **왔다갔다 여러 파일/함수를 오가며** 스택을 추적해야 해서 "지금 내가 어느 함수의 몇 번째 호출에 있는지" 헷갈리기 쉬워. 특히 3개 이상의 함수가 얽히면 순환 구조를 눈으로 따라가기가 급격히 어려워짐.

### 7. 유클리드 호제법이 빠른 이유

**결과**: `gcd(999999999, 987654321)`처럼 억 단위 숫자도 **재귀 호출 몇십 번**이면 끝나.

**(a)** 순진한 방법(1부터 min(x,y)까지 다 나눠보기)은 최악의 경우 `min(x,y)`번 나눗셈이 필요해 — 억 단위 숫자면 **억 번**. 유클리드 호제법은 몇십 번이면 끝나니 **수천만 배** 차이.

**(b)** `x % y`를 하면 나머지는 항상 `y`보다 작아. 그리고 실은 한 단계 지날 때마다 두 수 중 큰 쪽이 **최소 절반 이하로 줄어드는 경향**이 있어(피보나치수 같은 최악의 경우에도 로그 시간). 이게 7일차 이진 검색이 "매번 절반으로 줄여서" O(log n)이 되는 것과 **같은 원리** — 유클리드 호제법의 시간복잡도도 O(log(min(x,y)))에 가까워. "선형으로 하나씩 줄이기 vs 곱셈적으로 확 줄이기"의 차이.

### 8. 재귀로 리스트 합
```python
def list_sum(lst):
    if not lst:
        return 0
    else:
        return lst[0] + list_sum(lst[1:])
```
**왜**: `factorial(n) = n * factorial(n-1)`과 완전히 같은 패턴 — "리스트의 합 = 첫 원소 + (나머지 리스트의 합)". 빈 리스트(`not lst`가 참)가 기저 조건, `lst[1:]`로 매번 리스트가 하나씩 짧아지며 결국 빈 리스트에 도달.

### 9. 명시적 스택으로 재귀 흉내내기

- `stack.append(current)` 부분 = 재귀의 **"내려가기"** 단계. `factorial(4)`가 `factorial(3)`을 호출하며 "4를 곱해야 함"을 미뤄두는 것과 같아 — 여기선 그냥 `stack`에 4, 3, 2, 1을 직접 쌓아둠.
- `stack.pop()` 부분 = 재귀의 **"올라오기"** 단계. 기저 조건에서 반환된 값(1)에 스택에 쌓인 것들을 하나씩 곱하며 되짚어 올라가는 것과 같아.
- **왜 RecursionError 걱정이 없나**: 우리가 `while` 루프로 스택을 직접 관리하니까, 파이썬의 **함수 호출 콜 스택을 안 씀**. 대신 `stack`이라는 **일반 리스트(힙 메모리)**에 쌓기 때문에, 메모리가 허용하는 한 얼마든지 커질 수 있어. 이게 "재귀를 반복문 + 명시적 스택으로 바꾸면 RecursionError를 피할 수 있다"는 실전 테크닉의 원리야.

## 📌 핵심 3줄 요약

1. **재귀 = 자기 자신을 포함해서 정의**. 기저 조건(예: `n==0`)이 반드시 있어야 하고, 없으면 무한 재귀 → **RecursionError**.
2. 재귀 호출은 11일차 **콜 스택**에 쌓였다가(LIFO) 기저 조건에서 되짚어 풀린다. 반복문은 스택을 안 쓰니까 깊이 제한이 없다.
3. **직접 재귀**(자기 자신 호출) vs **간접 재귀**(여러 함수를 거쳐 순환) — 간접 재귀는 여러 함수를 오가며 추적해야 해서 디버깅이 더 어렵다.
