오늘 함수 이름을 `max`로 지었지. 이게 왜 위험한지 재현해봐.

````python
print(max([3, 1, 2]))   # 내장 max: 리스트 최댓값 → 잘 됨

def max(a, b, c):       # ← 내장 max를 덮어씀
    m = a
    if b > m: m = b
    if c > m: m = c
    return m

print(max(3, 1, 2))     # 내가 만든 버전: 잘 됨
# print(max([3, 1, 2])) # 주석 풀면? 무슨 에러가 왜?

# 질문: 함수 이름을 뭐로 바꾸면 안전할까? (2가지 이상)
````

### Part 7. 알고리즘
**17** 🟢 —
```python
def min3(a, b, c):
    m = a
    if b < m: m = b
    if c < m: m = c
    return m
```

**18** 🟡🔴 —
```python
def median_sort(a, b, c):
    return sorted([a, b, c])[1]

def median_chain(a, b, c):
    if c <= a <= b or b <= a <= c: return a
    elif c <= b <= a or a <= b <= c: return b
    else: return c

def median_trick(a, b, c):
    return a + b + c - max(a, b, c) - min(a, b, c)
```

**19** 🔴 —
1. `median3a`는 첫 `if`에서 `a`와 `b`를 비교하고, `elif`에서 `b`와 `a`를 **또** 비교 → 같은 쌍 중복 검사.
2. 중첩 if는 `a >= b`가 참임을 알면 그 사실을 **재사용**해 `b`와 `c`만 더 봄 → 비교 최소.
3. 3개는 몇 번 비교로 끝(정렬은 오버헤드). 개수가 늘면 손으로 짠 분기는 확장 불가, 정렬은 `sorted(...)[len//2]`로 일반화.
