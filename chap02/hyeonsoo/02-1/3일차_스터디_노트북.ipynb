**개념 9** 🔴 — `True`(값 같음), `False`(다른 객체), `True`(c는 a와 같은 값), `True`(c=a라 같은 객체), `False`.
> `a`,`b`는 내용은 같지만 **각각 따로 만들어진 리스트**라 `id`가 달라 → `==`는 True, `is`는 False. `c=a`는 같은 객체를 가리켜서 `is`도 True.

**개념 10** 🟡 —
| 비교 | 판정 과정 | 결과 |
|---|---|---|
| `[1,2,3] == [1,2,3]` | 모든 원소 동일 | `True` |
| `[1,2,3] < [1,2,4]` | 셋째서 3<4 | `True` |
| `[1,2,3] < [1,2,3,5]` | 앞 다 같고 왼쪽이 더 짧음 → 짧은 쪽이 작음 | `True` |
| `[1,3] < [1,2,9,9]` | 둘째서 3 vs 2 → 3>2라 왼쪽이 큼 | `False` |

> 규칙: 맨 앞부터 비교, 다른 원소가 나오면 그걸로 판정. 앞이 다 같으면 **짧은 쪽이 작다**.

**개념 11** 🟢 — 거짓: `0, "", [], {}, None`. 참: `1, -1, "a", [0]`.
> `[0]`은 참(리스트 자체가 안 비었음). `-1`도 참(0만 거짓).

**개념 12** 🔴 —
1. `x=[]`: `if x:` → **거짓**(빈 리스트는 falsy) / `if x is not None:` → **참**(None은 아니니까).
2. `if x:`는 "None인지"에 더해 "비었는지"까지 검사해버려서, 빈 리스트/0을 None처럼 잘못 걸러냄.
3. `[0]`은 리스트 자체에 원소가 1개 있어 안 비었으므로 참. 안의 값(0)이 falsy인 것과 **컨테이너가 비었는지**는 별개.

**개념 13** 🟢 — `5`, `7`, `99`, `166`, `33.2`.

**구현 5** 🔴 — `[[0]*3]*3`은 **같은 행 리스트를 3번 참조** → 한 칸 바꾸면 다 바뀜(`[[1,0,0],[1,0,0],[1,0,0]]`). 고치기:
```python
grid_fixed = [[0]*3 for _ in range(3)]   # 각 행을 새로 생성
```

**구현 6** 🟢
```python
result = sorted(set(data))   # set으로 중복 제거 → sorted로 정렬
# [1, 2, 3, 5]
```

## 🎯 3일차 핵심 3줄
1. **리스트=가변, 튜플=불변** — 낱개 변수를 배열로 묶으면 반복·정렬·집계가 가능해진다.
2. **슬라이싱 `[i:j:k]`은 관대(끝 제외, 범위 넘어도 OK), 인덱싱은 엄격(넘으면 에러)**.
3. **`==`는 값(등가성), `is`는 객체(동일성)** — 그리고 `if x:`는 "값 유무"가 아니라 truthy 판정.
