- **R-2**: `[50,40,30,20,10]`, `[20,30,40]`.

## Part A. 개념

**개념 1** 🟢 — 1) `a: Sequence`=매개변수 타입 힌트("시퀀스를 기대"), `-> Any`=반환 타입 힌트("아무 타입이나 반환"). 2) **에러 안 남** — 어노테이션은 실행 시 검사되지 않음(힌트일 뿐). 3) 사람에게 문서 역할 + 타입 체커(mypy)/에디터가 경고를 띄워주는 용도. 강제 아님.

**개념 2** 🟢 — 함수 첫 줄의 `'''...'''`(삼중따옴표 문자열)은 **독스트링**. `max_of.__doc__` → 그 문장이 출력. `help(max_of)` → 함수 설명 문서로 보여줌. (어노테이션은 `:`/`->`로 타입을 붙인 것, 독스트링은 함수 첫 줄의 문자열.)

**개념 3** 🟡 — 1) 직접 실행 → `'__main__'`. 2) import → `'max'`(모듈명). 3) 가드가 없으면 import하는 순간 `max.py` 하단의 `input(...)` 데모가 **멋대로 실행**됨(원치 않는 입력창). 가드가 그걸 막아 "직접 실행 시에만" 데모가 돌게 함.

**개념 4** 🟡 — `__name__`=(ㄴ), `__file__`=(ㄱ), `__doc__`=(ㄷ), `__init__`(메서드)=(ㄹ), `__init__.py`=(ㅁ).

**개념 5** 🔴 — `x` → **5 그대로**(안 바뀜), `y` → **[1,2,3,999]**(바뀜).
> `int`는 불변: `n += 100`은 새 객체를 만들어 지역 `n`이 그걸 가리킴 → 원본 `x` 무관. `list`는 가변: `lst.append`는 원본 객체 내부를 직접 수정 → 바깥 `y`도 바뀜. (call by object reference)

**개념 6** 🟡 — `[4,3,2,1] True`(reverse는 in-place, id 유지). `b=[1,2,3,4]`(안 바뀜), `c=[4,3,2,1]`, `b is c` → `False`(다른 객체).

**개념 7** 🟡 — 1) `max_of`는 읽기만(`Sequence`), `reverse_array`는 원소를 **수정**해서 가변이 필요(`MutableSequence`). 2) 튜플 넘기면 `a[i]=...`에서 **TypeError**(튜플은 항목 대입 불가). 3) `max_of`는 인덱싱·비교만 하고 수정 안 해서, 불변인 튜플·문자열도 OK.

**개념 8** 🟢 — `(0,'A') (1,'B') (2,'C')`(언패킹 안 하면 튜플 통째). 그다음 `1: A` `2: B` `3: C`(start=1).

## 🎯 4일차 핵심 3줄
1. **배열 스캔 = 챔피언 갱신** (`range(1, len(a))`) — 최대·최소·역순의 공통 뼈대.
2. **인수 전달은 call by object reference** — 불변(int/str)은 원본 안 바뀌고, 가변(list)은 바뀐다. `-> None`(in-place) vs 새 객체 반환.
3. **어노테이션은 힌트지 강제 아님** / `if __name__=='__main__':`은 "직접 실행 시에만" 데모 실행 → 모듈 재사용의 핵심.
