## 📦 부록 — FixedStack 전체 코드

`fixed_stack.py`가 같은 폴더에 없다면 **이 셀을 먼저 실행**해.

````
from typing import Any

class FixedStack:
    '''고정 길이 스택 클래스'''

    class Empty(Exception):
        '''비어 있는 FixedStack에 pop 또는 peek할 때 내보내는 예외'''
        pass

    class Full(Exception):
        '''가득 찬 FixedStack에 push할 때 내보내는 예외'''
        pass

    def __init__(self, capacity: int = 256) -> None:
        self.stk = [None] * capacity   # 스택 본체
        self.capacity = capacity       # 스택의 크기
        self.ptr = 0                   # 스택 포인터

    def __len__(self) -> int:
        return self.ptr

    def is_empty(self) -> bool:
        return self.ptr <= 0

    def is_full(self) -> bool:
        return self.ptr >= self.capacity

    def push(self, value: Any) -> None:
        if self.is_full():
            raise FixedStack.Full
        self.stk[self.ptr] = value
        self.ptr += 1

    def pop(self) -> Any:
        if self.is_empty():
            raise FixedStack.Empty
        self.ptr -= 1
        return self.stk[self.ptr]

    def peek(self) -> Any:
        if self.is_empty():
            raise FixedStack.Empty
        return self.stk[self.ptr - 1]

    def clear(self) -> None:
        self.ptr = 0

    def find(self, value: Any) -> Any:
        for i in range(self.ptr - 1, -1, -1):   # 꼭대기 -> 바닥
            if self.stk[i] == value:
                return i
        return -1

    def count(self, value: Any) -> int:
        c = 0
        for i in range(self.ptr):               # 바닥 -> 꼭대기
            if self.stk[i] == value:
                c += 1
        return c

    def __contains__(self, value: Any) -> bool:
        return self.count(value) > 0

    def dump(self) -> None:
        if self.is_empty():
            print('스택이 비어 있습니다.')
        else:
            print(self.stk[:self.ptr])

print('FixedStack 준비 완료')
````

## 📖 오늘의 핵심 개념

### 스택(stack) = LIFO
데이터를 **임시 저장**하는 자료구조. 입출력 순서는 **후입선출(LIFO, Last In First Out)** — 가장 나중에 넣은 걸 가장 먼저 꺼냄.

- **푸시(push)**: 데이터를 넣기
- **팝(pop)**: 데이터를 꺼내기 (가장 최근 것)
- **꼭대기(top)**: 윗부분 (push/pop이 일어나는 곳)
- **바닥(bottom)**: 아랫부분

접시를 쌓았다 빼는 것처럼, **맨 위에서만** 넣고 뺀다.

### 스택을 구성하는 3요소
| 이름 | 의미 |
|---|---|
| `stk` | 스택 본체 (list 배열). `stk[0]`이 **바닥** |
| `capacity` | 스택 최대 크기 = `len(stk)` |
| `ptr` (스택 포인터) | **쌓여 있는 데이터 개수**. 비면 0, 가득 차면 capacity |

핵심: `ptr`은 "**다음에 푸시할 자리의 인덱스**"이자 "**현재 데이터 개수**"야. 가장 마지막에 푸시한 원소는 `stk[ptr-1]`.

### push / pop / peek 의 ptr 조작
```python
def push(self, value):
    if self.is_full(): raise Full
    self.stk[self.ptr] = value   # 자리에 넣고
    self.ptr += 1                # 포인터 1 증가

def pop(self):
    if self.is_empty(): raise Empty
    self.ptr -= 1                # 포인터 1 감소
    return self.stk[self.ptr]    # 그 자리 값 반환

def peek(self):                  # 꺼내지 않고 들여다보기만
    if self.is_empty(): raise Empty
    return self.stk[self.ptr - 1]  # ptr은 안 바뀜
```

### 예외 처리 (교재 보충수업 4-1, 4-2)
- `Empty`: 빈 스택에 pop/peek 할 때
- `Full`: 가득 찬 스택에 push 할 때
- 둘 다 `Exception`의 하위 클래스로 정의 (BaseException 아님 — 사용자 정의 예외는 Exception에서 파생하는 게 원칙)

### 검색: find / count / `__contains__`
- `find`: **꼭대기→바닥** 방향 선형 검색. 먼저 팝할 데이터를 우선하려고. 실패 시 -1
- `count`: 바닥→꼭대기로 개수 세기
- `__contains__`: `in` 연산자를 쓸 수 있게 함. `count > 0`

### 던더(dunder) 메서드
`__len__`, `__contains__`처럼 앞뒤로 언더스코어 2개(`__`) 붙은 특별한 메서드. **더블 언더스코어 = 던더**.
- `__len__` 정의하면 `len(obj)` 가능
- `__contains__` 정의하면 `x in obj` 가능

### deque로 구현 (보충수업 4-4)
`collections.deque`는 양쪽 끝에서 O(1)로 추가·삭제하는 자료구조. 이걸로 스택을 훨씬 간단히 구현 가능. 단, **자료구조를 배우려면 FixedStack처럼 직접 짜보는 게 중요**.

### 6. [예측] FixedStack.find vs Stack.find — 방향이 다르다 🔥

교재 160p: "검색은 **꼭대기 쪽에서 바닥 쪽으로** 선형 검색을 합니다."

- `FixedStack.find`는 `range(ptr-1, -1, -1)` → 꼭대기부터
- `Stack.find`(deque 버전)는 `self.__stk.index(value)` → deque.index는 **바닥(왼쪽)부터**

같은 데이터가 든 두 스택에서 같은 값을 찾으면 **다른 인덱스**가 나와. 예측하고 실행해봐.

````
from collections import deque

# FixedStack.find 방식 (꼭대기부터)
class FixedFind:
    def __init__(self, data):
        self.stk = list(data)
        self.ptr = len(data)
    def find(self, value):
        for i in range(self.ptr - 1, -1, -1):
            if self.stk[i] == value:
                return i
        return -1

# Stack.find 방식 (deque.index = 바닥부터)
class DequeFind:
    def __init__(self, data):
        self.stk = deque(data)
    def find(self, value):
        try:
            return self.stk.index(value)
        except ValueError:
            return -1

data = [1, 2, 3, 1, 5]   # 1이 인덱스 0과 3에 있음
f = FixedFind(data)
d = DequeFind(data)

print('데이터:', data, '(1이 인덱스 0, 3에 존재)')
print('FixedStack.find(1):', f.find(1))   # 꼭대기부터 -> ?
print('Stack.find(1)     :', d.find(1))   # 바닥부터 -> ?
````

**(b)** 교재가 "꼭대기부터 검색"이라고 한 의도를 생각하면, 어느 쪽이 스택의 취지에 더 맞을까? (힌트: 스택에서 "먼저 만나는" 데이터는?)

### 7. [실험] FixedStack vs deque — 가득 찼을 때 동작이 다르다

`FixedStack`은 가득 차면 `Full` 예외를 던져. 그런데 `deque(maxlen=n)`은 가득 찬 상태에서 push하면 **예외 없이 반대쪽 원소를 밀어내**. 직접 확인해봐.

````
from collections import deque

# FixedStack: 가득 차면 Full 예외
class FixedFull:
    class Full(Exception): pass
    def __init__(self, cap):
        self.stk = [None]*cap; self.capacity = cap; self.ptr = 0
    def push(self, v):
        if self.ptr >= self.capacity:
            raise FixedFull.Full
        self.stk[self.ptr] = v; self.ptr += 1

# deque(maxlen): 가득 차면 밀어냄
dq = deque([], maxlen=3)

print('=== deque(maxlen=3)에 1,2,3,4,5 push ===')
for v in [1, 2, 3, 4, 5]:
    dq.append(v)
    print(f'  push({v}) -> {list(dq)}')
print()

print('=== FixedFull(3)에 1,2,3,4,5 push ===')
fs = FixedFull(3)
for v in [1, 2, 3, 4, 5]:
    try:
        fs.push(v)
        print(f'  push({v}) -> {fs.stk[:fs.ptr]}')
    except FixedFull.Full:
        print(f'  push({v}) -> Full 예외!')
````

### 9. [설명] 왜 `ptr >= capacity`이고 `ptr == capacity`가 아닐까?

교재 156p 보충설명에 이런 얘기가 나와:

> `is_empty`, `is_full`을 `==` 대신 `<=`, `>=`로 정의한 이유가 있다.

```python
def is_empty(self): return self.ptr <= 0        # == 0 아님
def is_full(self):  return self.ptr >= self.capacity   # == capacity 아님
```

- `ptr`은 정상적으론 항상 `0 <= ptr <= capacity` 범위인데, 왜 굳이 `<=`/`>=`로 방어적으로 짰을까?
- 만약 어떤 버그로 `ptr`이 -1이나 capacity+1이 됐을 때, `==` 버전과 `<=`/`>=` 버전은 각각 어떻게 동작할까?

## ✅ 정답 & 해설

### R-1. 스택의 find는 어떤 검색?

- **6일차 선형 검색**이야. 배열을 한 칸씩 순차적으로 훑으며 값을 비교하니까 **O(n)**. (이진 검색은 정렬 전제가 필요한데 스택은 정렬돼 있지 않으니 못 씀)
- **꼭대기→바닥으로 스캔하는 이유**: 스택은 LIFO라서 **먼저 팝될 데이터가 꼭대기**에 있어. 검색도 "먼저 꺼낼 것부터" 찾는 게 스택의 사용 맥락에 맞아. 같은 값이 여러 개면 꼭대기에 가까운(먼저 팝될) 인덱스를 반환.

---
### R-2. find -1 예측
- `find(20)` → **3** (꼭대기부터 스캔, 인덱스 3의 20을 먼저 만남)
- `find(10)` → **4** (꼭대기 = 인덱스 4의 10)
- `find(99)` → **-1** (없음, 6일차 선형 검색 실패와 동일)

### 1. LIFO 추적 정답

| 연산 | ptr | 스택 내용 | 반환값 |
|---|---|---|---|
| push(5) | 1 | [5] | - |
| push(7) | 2 | [5,7] | - |
| push(3) | 3 | [5,7,3] | - |
| pop() | 2 | [5,7] | **3** |
| push(9) | 3 | [5,7,9] | - |
| peek() | 3 | [5,7,9] | **9** (ptr 안 바뀜) |
| pop() | 2 | [5,7] | **9** |

핵심: peek은 ptr을 안 건드리고, pop은 ptr을 줄여. 그래서 push(9) 후 peek→pop하면 둘 다 9.

---
### 2. push/pop 구현 정답
```python
def push(self, value):
    if self.is_full():
        raise MyStack.Full
    self.stk[self.ptr] = value   # 자리에 넣고
    self.ptr += 1                # 포인터 증가

def pop(self):
    if self.is_empty():
        raise MyStack.Empty
    self.ptr -= 1                # 포인터 감소
    return self.stk[self.ptr]    # 그 자리 반환
```
**순서 주의**: push는 "넣고 → 증가", pop은 "감소 → 반환". pop에서 `ptr -= 1`을 먼저 해야 `stk[ptr]`이 마지막 원소를 가리켜.

---
### 3. peek vs pop

- `pop`은 `ptr -= 1`로 **포인터를 줄여서** 그 데이터를 "꺼낸" 것으로 처리 → 다음에 접근하면 그 아래 데이터가 꼭대기가 됨.
- `peek`은 ptr을 안 건드리므로 **데이터는 그대로** 남아있고 값만 확인.
- `peek` 3번 연속 → 매번 **같은 값**(현재 꼭대기)을 반환.
- `pop` 3번 연속 → 꼭대기부터 **차례로 다른 값** 3개를 반환하고 스택은 3개 줄어듦.

비유: peek은 "맨 위 접시 쳐다보기", pop은 "맨 위 접시 집어들기".

---
### 4. push 반환값 함정

**(a)** `push`의 시그니처는 `def push(self, value: Any) -> None`. 명시적 `return`이 없으면 파이썬 함수는 **`None`을 반환**해. 그래서 `x = s.push(x)`를 하면 스택엔 값이 잘 들어가지만 **`x`에는 `None`이 대입**돼.

**(b)** 지금 test 코드에선 그 다음에 `x`를 안 쓰니까 **당장은 무해**해. 하지만 만약 push 후에 `print(f'{x}를 넣었습니다')` 같은 걸 했다면 "None을 넣었습니다"가 출력되는 버그가 돼. 습관적으로 반환값 없는 메서드의 결과를 대입하지 않는 게 안전. → 교재처럼 그냥 `s.push(x)`.

---
### 5. `__contains__`가 int

**(a)** `in` 연산자는 `__contains__`의 반환값을 **자동으로 `bool()`로 강제 변환**해. `count(1)`이 `2`를 반환해도 `bool(2)` → `True`, `count(9)`가 `0`이면 `bool(0)` → `False`. 그래서 `in`으로 쓰면 우연히 정상 동작.

**(b)** **약한 버그(코드 스멜)**야. 동작은 맞지만:
1. 어노테이션이 `-> bool`인데 실제론 `int`를 반환 → 거짓말하는 문서 (8~9일차 `-> int` 문제와 같은 결)
2. 누군가 `x.__contains__(v)`를 직접 호출하면 `True/False`가 아니라 `2/0`이 나와서 놀랄 수 있음

`return self.count(value) > 0`으로 명시하면 의도가 분명하고 어노테이션과도 일치. FixedStack이 이렇게 돼 있음.

---
### 6. find 방향 차이

**(a)** `data = [1,2,3,1,5]`에서 1은 인덱스 0(바닥)과 3(꼭대기 근처)에 있어.
- `FixedStack.find(1)` → **3**: `range(ptr-1, -1, -1)`로 꼭대기부터 내려오며 스캔 → 인덱스 3의 1을 먼저 만남
- `Stack.find(1)` → **0**: `deque.index`는 왼쪽(바닥)부터 스캔 → 인덱스 0의 1을 먼저 만남

같은 값이 여러 개일 때 **스캔 시작 방향이 반대**라 다른 인덱스를 반환.

**(b)** 스택의 취지엔 **FixedStack(꼭대기부터)이 더 맞아**. 스택에서 "먼저 만나는/먼저 꺼낼" 데이터는 꼭대기에 있으니까, "이 값이 스택 어디쯤 있나?"의 답도 꼭대기 기준이 자연스러워. 교재 160p가 명시적으로 "꼭대기 쪽부터"라고 한 이유야. deque 버전은 `.index`를 그냥 쓰다 보니 이 의도가 깨진 케이스 — 편의(내장 메서드)와 정확성(스택 의미론)의 트레이드오프.

> 💡 deque 버전에서 꼭대기부터 찾으려면 뒤집어서 인덱스를 보정해야 해서 오히려 복잡해져. "내장 자료구조를 빌려 쓰면 편하지만 세부 의미가 원래 의도와 어긋날 수 있다"는 좋은 예시.

---
### 7. Full 동작 차이

**(a)**
- **FixedStack**: "정원 초과는 오류다" 철학. 가득 차면 `Full` 예외로 **명시적으로 거부**. 데이터를 절대 잃지 않음.
- **deque(maxlen)**: "최근 N개만 유지한다" 철학. 가득 차면 **가장 오래된 걸 조용히 버리고** 새 걸 받음. 데이터 손실이 정상 동작.

**(b)**
- **최근 N개 히스토리**(방문 기록, undo 20개): `deque(maxlen)`이 딱. 오래된 건 자동으로 밀려나니 코드가 간단.
- **데이터를 잃으면 안 되는 스택**: `FixedStack`. 넘치면 예외로 잡아서 처리(재할당, 경고 등)할 수 있음. deque면 조용히 유실돼서 버그를 놓침.

핵심: "가득 참"을 **오류로 볼지 정상으로 볼지**가 선택 기준.

### 8. 괄호 검사기 정답
```python
def is_balanced(s: str) -> bool:
    stack = []
    for ch in s:
        if ch == '(':
            stack.append(ch)       # push
        elif ch == ')':
            if not stack:          # 닫는데 열린 게 없음
                return False
            stack.pop()            # 짝 맞춰 pop
    return len(stack) == 0         # 안 닫힌 게 남았으면 False
```
**왜 스택인가**: 여는 괄호를 쌓아두고, 닫는 괄호가 오면 "가장 최근에 연 것"과 짝지어야 해. 이게 정확히 **LIFO**. 다 봤을 때 스택이 비어야(모든 짝이 맞아야) True. 이건 컴파일러의 문법 검사, 수식 계산 등에 실제로 쓰이는 스택의 대표 응용이야.

**엣지 케이스**: 빈 문자열 `""`은 True (짝이 안 맞는 게 없으니까). `"())("` 는 세 번째 `)`에서 스택이 비어 False.

---
### 9. `>=` vs `==`

- **정상 범위**에선 `ptr`이 항상 `0 <= ptr <= capacity`라서 `==`든 `>=`든 결과가 같아.
- **버그로 범위를 벗어나면** 차이가 생겨:
  - `is_full`이 `ptr == capacity`인데 버그로 `ptr`이 `capacity+1`이 됐다면 → `== capacity`는 **False** → "안 찼네?"라며 또 push → `stk[capacity+1]`에 접근 → **IndexError**
  - `ptr >= capacity`면 `capacity+1`도 **True** → push를 막아줌 → 배열 범위 밖 접근 차단
- 마찬가지로 `is_empty`가 `<= 0`이면 `ptr`이 음수로 망가져도 pop을 막아줌.

**교훈**: "이 값은 절대 이 범위를 안 벗어날 거야"라고 가정하지 말고, **경계를 부등호로 방어적으로** 잡으면 다른 버그가 있어도 배열 범위 밖 접근이라는 2차 재앙은 막을 수 있어. (6일차 이진검색 `pl > pr`, 오픈주소법 경계 처리와 같은 "경계를 넉넉히 잡는" 방어적 사고)
