### R-2. 🔴 [디버깅] 보초법 함수의 미묘한 버그

아래는 6일차 보초법 함수를 누가 살짝 바꿔 쓴 거야. 겉보기엔 멀쩡한데 특정 경우에 틀려. 어떤 입력에서 문제가 생길지 찾아봐.

````
from typing import Any, Sequence
import copy

def seq_search_sentinel(seq: Sequence, key: Any) -> int:
    a = copy.deepcopy(seq)
    a.append(key)
    i = 0
    while True:
        if a[i] == key:
            break
        i += 1
    return i   # <- 여기, 6일차 원본과 뭐가 달라졌지?

# 이 두 케이스 결과를 예측하고 실행해서 비교해봐
print(seq_search_sentinel([6, 4, 3, 2, 1, 2, 8], 2))   # 원본이면 3
print(seq_search_sentinel([6, 4, 3, 2, 1, 2, 8], 99))  # 원본이면 -1 ... 이건?
````

## 📖 오늘의 핵심 개념

### 이진 검색 (binary search)
**전제**: 배열이 **정렬**(오름차순/내림차순)되어 있어야 함. 정렬 안 된 배열엔 못 씀.

**아이디어**: 배열 중앙 원소부터 확인하고, 찾는 값과 비교해서 **탐색 범위를 매번 절반으로** 줄인다.

- `pl`(왼쪽 끝), `pr`(오른쪽 끝), `pc`(중앙) = `(pl + pr) // 2`
- `a[pc] == key` → **찾음** (종료 조건 1)
- `a[pc] < key` → 찾는 값이 오른쪽에 있음 → `pl = pc + 1` (뒤쪽 절반으로)
- `a[pc] > key` → 찾는 값이 왼쪽에 있음 → `pr = pc - 1` (앞쪽 절반으로)

### 이진 검색의 종료 조건
1. `a[pc] == key` → 검색 성공
2. **검색 범위가 더 이상 없는 경우** (`pl > pr`) → 검색 실패

> ⚠️ **우리가 세 번 헷갈린 지점**: 종료 조건 2는 반드시 **`pl > pr`**이어야 해. `pr == pc - 1`이나 `pr > pc`처럼 `pc`(방금 본 위치)를 기준으로 판단하면, 어느 한쪽 분기(elif/else)를 탈 때마다 우연히 조건이 맞아떨어져서 **탐색 범위가 남았는데도 조기 종료**돼. `pl`과 `pr`끼리 비교해야 "더 볼 구간이 없다"를 정확히 판단할 수 있어.

### 복잡도 (complexity)
알고리즘 성능을 객관적으로 평가하는 기준. 오늘은 **시간 복잡도**만.

- **선형 검색**: 평균 n/2번 비교 → n에 비례 → **O(n)**
- **이진 검색**: 매번 범위 절반 → **O(log n)**

O 표기법은 **가장 차원이 높은 항만** 남긴다: `O(1) + O(n) + O(1) = O(max(1, n, 1)) = O(n)`. 상수배·낮은 차수 항은 다 버림.

### 1. [예측] 이진 검색 손으로 추적하기

배열 `[1, 3, 5, 7, 9, 11, 13]`에서 `key = 11`을 이진 검색해. 아래를 **실행 전에** 손으로 계산해봐:

- 1회차: `pl=0, pr=6` → `pc=?`, `a[pc]=?`, 어느 분기?
- 2회차: `pl=?, pr=?` → `pc=?`, `a[pc]=?`, 어느 분기?
- 3회차: ...
- 몇 번 비교하고 어느 인덱스에서 찾을까?

그 다음 실행해서 맞는지 확인:

````
from typing import Any, Sequence

def bin_search(a: Sequence, key: Any) -> int:
    pl = 0
    pr = len(a) - 1
    count = 0
    while True:
        pc = (pl + pr) // 2
        count += 1
        print(f'{count}회차: pl={pl}, pr={pr}, pc={pc}, a[pc]={a[pc]}')
        if a[pc] == key:
            print(f'-> 찾음! 인덱스 {pc}')
            return pc
        elif a[pc] < key:
            pl = pc + 1
        else:
            pr = pc - 1
        if pl > pr:
            break
    print('-> 검색 실패')
    return -1

bin_search([1, 3, 5, 7, 9, 11, 13], 11)
````

### 4. [디버깅] 종료 조건 버그 3종 세트 🔥🔥

아래 세 버전은 종료 조건만 다르고 나머지는 동일해. 각각 `[1, 3, 5, 7, 9]`에서 여러 값을 검색했을 때 어떤 게 정상이고 어떤 게 틀리는지 예측한 뒤 실행해봐. **틀리는 버전은 왜 틀리는지**도 설명해.

- 버전 A: `if pr == pc - 1: break`
- 버전 B: `if pr > pc: break`
- 버전 C: `if pl > pr: break`

````
from typing import Any, Sequence

def make_bin_search(term):
    def bin_search(a, key):
        pl, pr = 0, len(a) - 1
        while True:
            pc = (pl + pr) // 2
            if a[pc] == key:
                return pc
            elif a[pc] < key:
                pl = pc + 1
            else:
                pr = pc - 1
            if term == 'A' and pr == pc - 1:
                break
            if term == 'B' and pr > pc:
                break
            if term == 'C' and pl > pr:
                break
        return -1
    return bin_search

a = [1, 3, 5, 7, 9]
for term in ['A', 'B', 'C']:
    f = make_bin_search(term)
    results = [f(a, k) for k in [1, 3, 5, 7, 9]]
    print(f'버전 {term}: {results}  (정답이면 [0,1,2,3,4])')
````

### 7. [실험] 선형 vs 이진, 비교 횟수 직접 재보기

n을 키워가면서 선형 검색과 이진 검색의 **최악 비교 횟수**를 실제로 재봐. 이진 검색이 log n 곡선을 그리는지 눈으로 확인.

````
import math

def linear_count(a, key):
    count = 0
    for i in range(len(a)):
        count += 1
        if a[i] == key:
            return count
    return count

def bin_count(a, key):
    pl, pr = 0, len(a) - 1
    count = 0
    while True:
        pc = (pl + pr) // 2
        count += 1
        if a[pc] == key:
            return count
        elif a[pc] < key:
            pl = pc + 1
        else:
            pr = pc - 1
        if pl > pr:
            return count

print(f"{'n':>8} | {'선형(최악)':>10} | {'이진(최악)':>10} | {'log2(n)':>8}")
print('-' * 48)
for n in [10, 100, 1000, 10000, 100000]:
    a = list(range(n))
    # 최악: 맨 끝 값 검색 (선형은 n번, 이진은 log n번)
    lin = linear_count(a, n - 1)
    bn = bin_count(a, n - 1)
    print(f'{n:>8} | {lin:>10} | {bn:>10} | {math.log2(n):>8.2f}')
````

## ✅ 정답 & 해설

### R-1. 보초법 vs 이진검색

- **보초법**: 배열 끝에 찾는 값(보초)을 심어서, 매 반복의 종료 조건 체크를 2개→1개로 줄여. 하지만 여전히 원소를 **처음부터 하나씩** 다 보니까 비교 횟수 자체는 그대로 → **O(n)**. "비교 1번당 드는 비용"을 깎는 최적화.
- **이진검색**: 매번 범위를 절반씩 버리니까 비교 **횟수 자체**가 log n으로 확 줄어 → **O(log n)**. "비교 횟수"를 깎는 최적화.
- **전제 조건 차이**: 이진검색은 "중앙값보다 크면 오른쪽"이라고 확신하려면 배열이 **정렬**돼 있어야 해. 보초법은 순서와 무관하게 그냥 훑는 거라 정렬이 필요 없어.

### R-2. 보초법 버그

**문제**: 마지막 줄이 `return i`로 바뀌었어. 원본은 `return -1 if i == len(seq) else i`.

**결과**: 배열에 없는 값(99)을 검색하면, 보초 위치(`i == len(seq)`)에서 멈추는데 그걸 그대로 `return i` 해버려서 **-1이 아니라 원소 개수(7)**를 반환해. "찾았다"고 착각하게 만드는 버그.

**왜**: 보초법의 핵심은 "멈춘 위치가 보초 자리면 = 원본엔 없던 값 = 실패"를 판별하는 거야. 그 판별(`i == len(seq)` 체크)을 빼먹으면 보초를 진짜 데이터로 오인해.

### 3. 정렬이 필요한 이유

이진 검색의 핵심 논리는 "`a[pc] < key`면 찾는 값은 반드시 `pc`보다 **오른쪽**에 있다"야. 이 확신은 배열이 정렬돼 있어서 **`pc` 왼쪽은 전부 `a[pc]` 이하**라는 게 보장될 때만 성립해. 정렬이 안 돼 있으면 `a[pc]`가 key보다 작아도 찾는 값이 왼쪽에 있을 수 있어서, 절반을 버리는 순간 답을 놓칠 수 있어. → 그래서 정렬이 필수 전제.

### 4. 디버깅 3종 세트 🔥
- **버전 A** (`pr == pc - 1`): 틀림. `else` 분기(`pr = pc - 1`)를 타는 순간 방금 만든 값이라 항상 참 → 왼쪽 절반을 다 못 보고 조기 종료.
- **버전 B** (`pr > pc`): 틀림. `elif` 분기(`pl = pc + 1`)를 타면 `pr`은 안 바뀌고 그대로 남아서 `pr > pc`가 자주 참 → 오른쪽 절반에서 조기 종료.
- **버전 C** (`pl > pr`): **정답**. 분기와 무관하게 "왼쪽 경계가 오른쪽 경계를 넘음 = 더 볼 구간 없음"을 정확히 판단.

**교훈**: 종료 조건은 항상 탐색 경계 두 개(`pl`, `pr`)끼리 비교. 방금 본 위치(`pc`)를 끌어들이면 분기에 따라 우연히 참/거짓이 갈려서 불안정.

### 6. O(n) 이유

O 표기법은 n이 무한대로 커질 때의 **증가율**만 본다. 상수 항 `O(1)`들은 n이 아무리 커져도 그대로지만, `O(n)` 항은 n에 비례해 계속 커져. 그러니 n이 충분히 크면 전체 실행 시간을 지배하는 건 `O(n)` 항이고, 나머지는 무시해도 될 만큼 작아져 → `O(max(1, n, 1)) = O(n)`.

---
### 7. 선형 vs 이진 실험 해설
선형은 n에 정비례(10→10, 100→100, ... 100000→100000)로 폭증하지만, 이진은 log 곡선이라 n이 10배가 돼도 비교 횟수는 겨우 3~4번씩만 늘어(약 log₂10 ≈ 3.3). n=100000일 때 선형 100000번 vs 이진 17번. 이게 O(n)과 O(log n)의 실전 차이.

### 8. leftmost 이진 검색 정답

````
from typing import Any, Sequence

def bin_search_leftmost(a: Sequence, key: Any) -> int:
    pl, pr = 0, len(a) - 1
    result = -1
    while pl <= pr:
        pc = (pl + pr) // 2
        if a[pc] == key:
            result = pc      # 후보로 기록
            pr = pc - 1      # 더 왼쪽 탐색 계속
        elif a[pc] < key:
            pl = pc + 1
        else:
            pr = pc - 1
    return result

print(bin_search_leftmost([1, 2, 2, 2, 3, 4], 2))   # 1
print(bin_search_leftmost([1, 2, 3, 4, 5], 3))      # 2
print(bin_search_leftmost([1, 2, 3, 4, 5], 9))      # -1
````

**왜**: 값을 찾아도 즉시 멈추지 않고, "일단 후보로 저장 + 왼쪽을 더 파본다"를 반복해. 왼쪽에 같은 값이 더 없을 때까지 `pr`을 계속 당기니까, 최종적으로 가장 작은 인덱스가 `result`에 남아. (`while pl <= pr` 구조를 쓰면 break 종료 조건 실수도 원천 차단돼서 이런 변형에 더 안전해.)

---
### 9. 오버플로 해설

- **파이썬은 안전**: 파이썬 정수(int)는 임의 정밀도(arbitrary precision)라서 크기 제한이 없어. `pl + pr`이 아무리 커도 오버플로가 안 나. C/Java는 int가 고정 비트(보통 32/64비트)라 최대값을 넘으면 음수로 뒤집혀서 버그가 남.
- **언어 무관 안전한 방법**: `pc = pl + (pr - pl) // 2`. `pr - pl`은 항상 배열 크기 이하라 절대 오버플로 안 나고, 결과는 `(pl + pr) // 2`와 동일해. 습관으로 이 형태를 쓰면 어느 언어로 옮겨도 안전.

## 📌 핵심 3줄 요약

1. **이진 검색**은 정렬된 배열에서 매번 범위를 절반씩 버려 O(log n)에 검색한다. 종료 조건은 반드시 **`pl > pr`** (경계끼리 비교, `pc`를 끌어들이면 조기 종료 버그).
2. **선형 O(n) vs 이진 O(log n)**: n=100000이면 100000번 vs 17번. 대신 이진은 정렬이라는 전제가 필요.
3. **O 표기법**은 n이 커질 때 지배하는 최고차항만 남긴다: `O(1)+O(n)+O(1)=O(n)`.
